```python
video_path=r'C:\Users\singh\open_cv\New folder\videos\VID_20230613_153837.mp4'
```

```python
# to capture video
cap=cv2.VideoCapture(video_path)

while cap.isOpened():
    raju,frame = cap.read()
    if raju:
        frame = cv2.resize(frame,(600, 400))
        cv2.imshow('video_play',frame)
        
        if cv2.waitKey(25) & 0xff==ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

# DIFFERENT THINGS IN DIFFERENT FRAME

```python
import cv2
import numpy as np

path = r'''C:\Users\singh\open_cv\New folder\photos'''
img_name = "licensed-image.jpeg"

cap1 = cv2.VideoCapture(video_path)  # babu's video
cap2 = cv2.VideoCapture(0)  # camera

while cap1.isOpened() and cap2.isOpened():  # Check if both video streams are open
    raju1, frame1 = cap1.read()
    raju2, c_frame = cap2.read()

    if raju1 and raju2:  # Check if frames were successfully read from both streams
        frame1 = cv2.resize(frame1, (600, 400))
        c_frame = cv2.resize(c_frame, (600, 400))

        image = cv2.imread(path + '\\' + img_name)
        image = cv2.resize(image, (600, 400))

        frame_2 = np.hstack((image, c_frame))
        frame_02 = np.hstack((frame1, frame1))  # Use the correct frame variable
        frame_4 = np.vstack((frame_2, frame_02))

        cv2.imshow('video_play', frame_4)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap1.release()
cap2.release()
cv2.destroyAllWindows()
```

cv2.imread(path,1) or cv2.imread(path,cv2.IMREAD_COLOR) -- reads the image with RGB

cv2.imread(path,-1) or cv2.imread(path,cv2.IMREAD_UNCHANGED)--it reads the image as is form the source.if the source image is RGB, it loads the image into array with red,green,blue channels.If the source image is ARGB, it loads the image with three colors components along with the alpha or transparency channel

# save image from video

```python
video_path=r'C:\Users\singh\open_cv\New folder\videos\VID_20230613_153837.mp4'

os.mkdir('video_to_image')

cap=cv2.VideoCapture(video_path)

img_count = 1

while cap.isOpened():
    raju,frame = cap.read()
    
    if not raju:
        print('unable to read frame')
        break
        
    is_img_write=cv2.imwrite(f'video_to_image\image{img_count}.png',frame)
    
    if is_img_write:
        print(f'video_to_image\image{img_count}.png')
        
        
        cv2.imshow('video_play',frame)
        
        cv2.waitKey(25) & 0xff==ord('q')
        
        img_count+=1
            
            

cap.release()
cv2.destroyAllWindows()
```
